# Epsilon Decay

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#
# Define Constants
#
N_steps=10000 # number of steps
```

```python
def plot_epsilon(N_steps):
    #
    # Plots epsilon over a specified number of steps (num_steps) for:
    # ---------------------------------------------------------------
    #  1) Epsilon = 0.1 (no decay)
    #  2) Epsilon = 1.0, decay=0.9
    #  3) Epsilon = 1.0, decay=0.99
    #  4) Epsilon = 1.0, decay=0.999
    #  5) Epsilon = 1.0, decay=0.9999
    # ---------------------------------------------------------------
    #

    # Case (a): Epsilon = 0.1, no decay
    eps_a = [0.1 for _ in range(N_steps)]

    # Case (b): Epsilon = 1.0, decay rate = 0.9
    decay_b = 0.9
    eps_b = [1.0 * (decay_b**i) for i in range(N_steps)]

    # Case (c): Epsilon = 1.0, decay rate = 0.99
    decay_c = 0.99
    eps_c = [1.0 * (decay_c**i) for i in range(N_steps)]

    # Case (d): Epsilon = 1.0, decay rate = 0.999
    decay_d = 0.999
    eps_d = [1.0 * (decay_d**i) for i in range(N_steps)]

    # Case (e): Epsilon = 1.0, decay rate = 0.9999
    decay_e = 0.9999
    eps_e = [1.0 * (decay_e**i) for i in range(N_steps)]

    # Plotting
    plt.figure(figsize=(10,6))
    plt.plot(eps_a, label='Eps=0.1 (No Decay)')
    plt.plot(eps_b, label='Eps=1.0, Decay=0.9')
    plt.plot(eps_c, label='Eps=1.0, Decay=0.99')
    plt.plot(eps_d, label='Eps=1.0, Decay=0.999')
    plt.plot(eps_e, label='Eps=1.0, Decay=0.9999')
    plt.xlabel('Steps')
    plt.ylabel('Epsilon')
    plt.title('Epsilon Decay Comparison')
    plt.legend()
    plt.show()
```

```python
# Call the plotting function
plot_epsilon(N_steps)
```

  QN_1:
  
  From the plots, what can we say about the agents ability to find a random arm
        in later stage of the 10,000 steps in cases (d) and (e) comparing with cases (b) and (c)?

  ANS:

      In cases (d) and (e), the decay rate is very slow, so the epsilon stays higher for longer.
      This means the agent continues to explore for a much larger portion of the 10,000 steps.
      As a result, it has a better chance of finding a random arm later in training.
       
      In contrast, in cases (b) and (c), the decay rate is faster, so the epsilon drops
      at a much faster rate. As a result, the agent will shift to exploiting earlier, making it
      less likely to explore new or under-sampled arms in the later stages. This can be good if the agent
      has already found a strong arm, but it may miss out on discovering a better arm that shows up
      later.

 QN_2:

 How would the values of the average total reward in cases (d) and (e)
       compared with cases (b) and (c)? Explain your expectation.

 ANS:  

       In cases (d) and (e), where the decay rate is very slow, the agent retains a high level of exploration
       throughout a significant portion of the 10,000 steps. This prolonged exploration would result in the agent
       sampling more random arms rather than fully committing to the optimal arm. Consequently, the agent in these
       cases (in the template) would likely lead to a lower overall average total reward unless majority of the arms have a high probability
       of reward which is unlikely.

       In contrast, in cases (b) and (c), where the decay is much faster, the agent quickly reduces its exploration rate
       and begins to exploit the best arm almost immediately. This early commitment ensures that the majority of the 10,000
       steps are spent pulling the optimal arm, resulting in a higher average total reward for the template cases.
